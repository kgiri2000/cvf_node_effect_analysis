```python
import pandas as pd
import seaborn as sns

from matplotlib import pyplot as plt
```

```python
folder = "../results/"
graphs = "graph_7_nodes"

```

```python
def get_df(graph):
    df = pd.read_csv(f"{folder}rank_effect_{graphs}.csv")
    df['CVF (Avg)'] = df['CVF In (Avg)'] + df['CVF Out (Avg)']
    df['CVF (Max)'] = df['CVF In (Max)'] + df['CVF Out (Max)']
    return df

    
```

```python
def plot_pt_rank_effect_avg(df, ax):
    sns.lineplot(data = df, x = "Rank Effect", y = 'PT (Avg)', ax = ax)
    ax.set(xlabel = f"Rank Effect", ylabel = "Count")
    ax.set_title("Program Transition Avg Effect")
    ax.set_yscale('log')

def plot_pt_rank_effect_max(df, ax):
    sns.lineplot(data = df, x = "Rank Effect", y = 'PT (Max)', ax = ax)
    ax.set(xlabel = f"Rank Effect", ylabel = "Count")
    ax.set_title("Program Transition Max Effect")
    ax.set_yscale('log')

def plot_rank_effect_cvf_avg(df, ax):
    sns.lineplot(data=df, x='Rank Effect', y='CVF (Avg)', ax=ax)
    ax.set(xlabel=f"Rank Effect", ylabel="Count")
    ax.set_title("CVF Avg")
    ax.set_yscale('log')

def plot_rank_effect_cvf_max(df, ax):
    sns.lineplot(data=df, x='Rank Effect', y='CVF (Max)', ax=ax)
    ax.set(xlabel=f"Rank Effect", ylabel="Count")
    ax.set_title("CVF Max")
    ax.set_yscale('log')
```

```python
# Plotting in a 2x2 grid
def plot_all(graph):
    df = get_df(graph)
    
    fig, axs = plt.subplots(2, 2, figsize=(20, 12), constrained_layout = True)

    plot_pt_rank_effect_avg(df, axs[0, 0])
    plot_pt_rank_effect_max(df, axs[0, 1])
    plot_rank_effect_cvf_avg(df, axs[1, 0])
    plot_rank_effect_cvf_max(df, axs[1, 1])

    plt.tight_layout()
    plt.show()
plot_all('graphs')
```
