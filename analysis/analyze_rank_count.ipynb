```python
import pandas as pd
import seaborn as sns

from matplotlib import pyplot as plt
```

```python
folder = "../results/"
graph = "graph_7_nodes"
```

```python
data = pd.read_csv(f"{folder}rank_{graph}.csv")
```

```python
# Create a line plot
sns.lineplot(x='Rank', y='Count (Max)', data=data, marker='o', label='Count (Max)')
sns.lineplot(x='Rank', y='Count (Avg)', data=data, marker='o', label='Count (Avg)')
# Add labels and title
plt.xlabel('Rank')
plt.ylabel('Count')
plt.title('Count (Max) and Count (Avg) vs. Rank')

# Show legend
plt.legend()

# Show the plot
plt.show()
```
